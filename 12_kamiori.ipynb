```python
#問題5
#for文でリストに紙を43回折るときの厚さの過程の値合計44個を記録するコード
THICKNESS = 0.00008 #折る前の紙の厚さ
number = 43 #紙を折る回数
folded_thickness_for = THICKNESS
folded_thickness_list = [folded_thickness_for]
for i in range(number):
    folded_thickness_for *= 2
    folded_thickness_list.append(folded_thickness_for)
print(folded_thickness_list)
print(len(folded_thickness_list))
```

```python
#問題6 練習
#紙を43回折った時の折れ線グラフの表示
import matplotlib.pyplot as plt
import japanize_matplotlib

x = list(range(44))
y = folded_thickness_list

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("折った回数")
ax.set_ylabel("紙の厚さ")
plt.show()
```
